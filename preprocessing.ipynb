```python
import pandas as pd
import os
import glob
```

### Filtra dados úteis dos estabelecimentos e coloca no padrão para leitura

```python
paths = ['data/Estabelecimentos{i}/*'.format(i=i) for i in range(10)]

for i, path in enumerate(paths):
    
    files = glob.glob(path)
    print(files)

    df = pd.read_csv("{file}".format(path=path, file=files[0]), sep=';',encoding = "ISO-8859-1", header=None)

    df.columns = ['CNPJ BÁSICO', 'CNPJ ORDEM', 'CNPJ DV', 'IDENTIFICADOR MATRIZ/FILIAL', 'NOME FANTASIA', 'SITUAÇÃO CADASTRAL', 
                                                                                                                     'DATA SITUAÇÃO CADASTRAL', 'MOTIVO SITUAÇÃO CADASTRAL', 'NOME DA CIDADE NO EXTERIOR', 'PAIS', 'DATA DE INÍCIO ATIVIDADE', 'CNAE FISCAL PRINCIPAL', 
                                                                                                                     'CNAE FISCAL SECUNDÁRIA', 'TIPO DE LOGRADOURO', 'LOGRADOURO', 'NÚMERO', 'COMPLEMENTO', 'BAIRRO', 'CEP', 'UF', 'MUNICÍPIO', 'DDD 1', 'TELEFONE 1','DDD 2', 'TELEFONE 2', 
                                                                                                                     'DDD DO FAX', 'FAX', 'CORREIO ELETRÔNICO', 'SITUAÇÃO ESPECIAL', 'DATA DA SITUAÇÃO ESPECIAL']
    colunas_uteis = ['CNPJ BÁSICO', 'CNPJ ORDEM', 'CNPJ DV', 'IDENTIFICADOR MATRIZ/FILIAL', 'SITUAÇÃO CADASTRAL', 'DATA SITUAÇÃO CADASTRAL', 'DATA DE INÍCIO ATIVIDADE', 'CNAE FISCAL PRINCIPAL', 'UF']

    df[df['UF']=='MG'][colunas_uteis].to_csv("data/Estabelecimentos_MG{i}.csv".format(i=i), index=False)
```

#### Função para converter e filtrar outros dados 


```python
def convert_files(paths, df_columns, useful_columns, output_name):

    for i, path in enumerate(paths):

        files = glob.glob(path)
        print(files)

        df = pd.read_csv("{file}".format(path=path, file=files[0]), sep=';',encoding = "ISO-8859-1", header=None)

        df.columns = df_columns

        columns_uteis = useful_columns
        df[columns_uteis].to_csv("data/{output_name}{i}.csv".format(output_name=output_name,i=i), index=False)

convert_files(paths=['data/Empresas{i}/*'.format(i=i) for i in range(10)], df_columns=['cnpj', 'razao social', 'NATUREZA', 'QUALIFICAÇÃO', 'CAPITAL', 'PORTE', 'ENTE'], useful_columns=['cnpj', 'NATUREZA', 'QUALIFICAÇÃO', 'CAPITAL', 'PORTE'], output_name='Empresas')
```

```python
convert_files(paths=['data/Naturezas/*'], df_columns=['codigo', 'Natureza Juridica'], useful_columns=['codigo', 'Natureza Juridica'], output_name='Naturezas')
```

```python
convert_files(paths=['data/Cnaes/*'], df_columns=['CODIGO', "NOME DA ATIVIDADE ECONÔMICA"], useful_columns=['CODIGO', "NOME DA ATIVIDADE ECONÔMICA"], output_name='Cnaes')
```

```python
convert_files(paths=['data/Qualificacoes/*'], df_columns=['CODIGO', 'NOME DA QUALIFICAÇÃO DO SÓCIO'], useful_columns=['CODIGO', 'NOME DA QUALIFICAÇÃO DO SÓCIO'], output_name='Qualificacoes')
```

Processa dados das empresas

É necessário juntar os 10 arquivos em um só

```python
def concat_files(files_name, output_name):
    df = pd.read_csv(files_name[0], sep=',')

    # Loop por todos os arquivos CSV restantes e os concatena ao dataframe df
    for file in files_name[1:]:
        print(file)
        df = pd.concat([df, pd.read_csv(file, sep=',')], ignore_index=True)

    # Salva o dataframe resultante como um arquivo CSV
    df.to_csv('data/{output_name}.csv'.format(output_name=output_name), index=False)
files= ['data/Empresas0.csv',
 'data/Empresas1.csv',
 'data/Empresas2.csv',
 'data/Empresas3.csv',
 'data/Empresas4.csv',
 'data/Empresas5.csv',
 'data/Empresas6.csv',
 'data/Empresas7.csv',
 'data/Empresas8.csv',
 'data/Empresas9.csv']
concat_files(files, 'Empresas')
```

```python
files= ['data/Estabelecimentos_MG0.csv',
 'data/Estabelecimentos_MG1.csv',
 'data/Estabelecimentos_MG2.csv',
 'data/Estabelecimentos_MG3.csv',
 'data/Estabelecimentos_MG4.csv',
 'data/Estabelecimentos_MG5.csv',
 'data/Estabelecimentos_MG6.csv',
 'data/Estabelecimentos_MG7.csv',
 'data/Estabelecimentos_MG8.csv',
 'data/Estabelecimentos_MG9.csv']

concat_files(files, 'Estabelecimentos')
```

Com os arquivos devidamente convertidos e tratados podemos realizar a leitura 

```python
df_estab = pd.read_csv("data/Estabelecimentos.csv", sep=',')
df_empresas = pd.read_csv("data/Empresas.csv", sep=',')
df_naturezas = pd.read_csv("data/Naturezas0.csv", sep=',')
df_cnaes = pd.read_csv("data/Cnaes0.csv", sep=',')
```

```python
df_estab	
```

```python
df_empresas_estab_mg = df_empresas.merge(df_estab, left_on="cnpj", right_on="CNPJ BÁSICO", how='inner')
df_empresas_estab_mg = df_empresas_estab_mg.drop_duplicates(subset=['cnpj'], keep='first')
df_empresas_estab_mg['DATA DE INÍCIO ATIVIDADE'] = df_empresas_estab_mg['DATA DE INÍCIO ATIVIDADE'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
df_empresas_estab_mg['DATA SITUAÇÃO CADASTRAL'] = df_empresas_estab_mg['DATA SITUAÇÃO CADASTRAL'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d', errors='coerce'))
```
